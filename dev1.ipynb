Convex hull specified geometrification:

```python
collisions.head()
```

```python
nullities = dict()
for identifier, geo_group in collisions.groupby("BOROUGH"):
    point_groups = geo_group.groupby('LOCATION')
    counts = point_groups.count().apply(lambda srs: srs.sum(), axis='columns')
    entries = point_groups.apply(len)
    width = len(geo_group.columns)
    geographic_nullity = np.average(1 - counts / width / entries)
    points = [eval(g)[::-1] for g in counts.index.values]
    hull = shapely.geometry.MultiPoint(points).convex_hull
    nullities[identifier] = {'nullity': geographic_nullity, 'shape': hull}
    print(geographic_nullity)
```

```python
nullities
```

```python
gpd.GeoSeries([nullities[key]['shape'] for key in nullities.keys()]).plot()
mplleaflet.show()
```

At this point, figure out how to plot without using `geopandas` as an intermediary and slap a colormap into the resultant `matplotlib` plot, and we're done. This looks promising, though imperfect...

```python
nullities = dict()
for identifier, geo_group in collisions.groupby("ZIP CODE"):
    point_groups = geo_group.groupby('LOCATION')
    counts = point_groups.count().apply(lambda srs: srs.sum(), axis='columns')
    entries = point_groups.apply(len)
    width = len(geo_group.columns)
    geographic_nullity = np.average(1 - counts / width / entries)
    points = [eval(g)[::-1] for g in counts.index.values]
    hull = shapely.geometry.MultiPoint(points).convex_hull
    nullities[identifier] = {'nullity': geographic_nullity, 'shape': hull}
    print(geographic_nullity)
```

```python
nullities
```

```python
gpd.GeoSeries([nullities[key]['shape'] for key in nullities.keys()]).plot()
mplleaflet.show()
```
